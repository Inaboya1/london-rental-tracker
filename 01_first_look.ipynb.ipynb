```python
# outcodes with fewer than 100 properties
outcode_counts = df['outcode'].value_counts()
small_outcodes = outcode_counts[outcode_counts < 100]
print(
    f"outcodes with fewer than 100 properties: {len(small_outcodes)}"
)

print(small_outcodes)
```

```python
# Create a flag for whether an outcode is well-represented
# We will use this later to filter out unreliable averages
reliable_outcodes = outcode_counts[outcode_counts >= 100].index
df['outcode_reliable'] = df['outcode'].isin(reliable_outcodes)
print(f'Properties in reliable outcodes: {df["outcode_reliable"].sum():,}')
```

```python
# Convert the three date columns from text to real datetime objects
df['estimate_date'] = pd.to_datetime(df['estimate_date'], errors='coerce')
df['last_sale_date'] = pd.to_datetime(df['last_sale_date'], errors='coerce')
df['hist_date'] = pd.to_datetime(df['hist_date'], errors='coerce')
# errors='coerce' is a safety net
# If a date value is broken or empty, it becomes NaT (Not a Time)
# instead of crashing your whole notebook
# Verify the conversion worked
print('Date column types after conversion:')
print(df[['estimate_date','last_sale_date','hist_date']].dtypes)
# These should now show: datetime64[ns]
# Now you can extract useful information from dates
df['hist_year'] = df['hist_date'].dt.year
df['hist_month'] = df['hist_date'].dt.month
# How many sales per year?
print('\nSales by year in history:')
print(df['hist_year'].value_counts().sort_index())
# What is the earliest and most recent sale date?
print(f'\nEarliest sale in history: {df["hist_date"].min().date()}')
print(f'Most recent sale in history: {df["hist_date"].max().date()}')
```

```python
# Median rent by outcode
rent_data = df.dropna(subset=['rent']) # only rows with rent values
# Group by outcode and calculate median rent + how many properties
rent_by_outcode = (rent_data
.groupby('outcode')['rent']
.agg(median_rent='median', count='count')
.reset_index())
# Only keep outcodes with at least 100 properties — small samples mislead
rent_by_outcode = rent_by_outcode[rent_by_outcode['count'] >= 100]
# Sort by median rent — most expensive at the top
rent_by_outcode = rent_by_outcode.sort_values('median_rent', ascending=False)
# Print top 15 most expensive and bottom 15 most affordable
print('Top 15 most expensive outcodes to rent:')
print(rent_by_outcode.head(15).to_string(index=False))
print('\nBottom 15 most affordable outcodes:')
print(rent_by_outcode.tail(15).to_string(index=False))
# Chart — top 20 and bottom 20 side by side
top20 = rent_by_outcode.head(20)
bottom20 = rent_by_outcode.tail(20)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
ax1.barh(top20['outcode'][::-1], top20['median_rent'][::-1],
color='#993C1D', edgecolor='white')
ax1.set_title('20 Most Expensive Outcodes\n(median monthly rent)',
fontsize=11, pad=10)
ax1.set_xlabel('Median monthly rent (£)')
ax1.grid(True, axis='x', alpha=0.3)
ax2.barh(bottom20['outcode'][::-1], bottom20['median_rent'][::-1],
color='#0F6E56', edgecolor='white')
ax2.set_title('20 Most Affordable Outcodes\n(median monthly rent)',
fontsize=11, pad=10)
ax2.set_xlabel('Median monthly rent (£)')
ax2.grid(True, axis='x', alpha=0.3)
plt.suptitle('London Rental Market by Outcode', fontsize=14, y=1.01)
plt.tight_layout()
plt.savefig('outputs/09_rent_by_outcode.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ANALYSIS 2: Gross rental yield by outcode
yield_data = df.dropna(subset=['rent','sale_price']).copy()
# Calculate gross annual yield: (annual rent / sale price) x 100
yield_data['annual_rent'] = yield_data['rent'] * 12
yield_data['gross_yield'] = (yield_data['annual_rent'] /
yield_data['sale_price'] * 100).round(2)
# Remove unrealistic yields — below 1% or above 15% are likely data errors
yield_data = yield_data[yield_data['gross_yield'].between(1, 15)]
# Median yield by outcode (minimum 50 properties for reliability)
yield_by_outcode = (yield_data
.groupby('outcode')['gross_yield']
.agg(median_yield='median', count='count')
.query('count >= 50')
.sort_values('median_yield', ascending=False)
.reset_index()
)
print('Top 10 highest-yielding outcodes:')
print(yield_by_outcode.head(10).to_string(index=False))
fig, ax = plt.subplots(figsize=(12, 6))
top_yield = yield_by_outcode.head(20)
ax.bar(top_yield['outcode'], top_yield['median_yield'],
color='#4B3FA8', edgecolor='white')
ax.axhline(yield_by_outcode['median_yield'].median(),
color='#BA7517', linestyle='--', linewidth=2,
label=f'Overall median yield')
ax.set_title('Gross rental yield by outcode — top 20', fontsize=13, pad=12)
ax.set_xlabel('Outcode', fontsize=10)
ax.set_ylabel('Gross annual yield (%)', fontsize=10)
ax.tick_params(axis='x', rotation=45)
ax.legend()
ax.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('outputs/10_rental_yield_by_outcode.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ANALYSIS 3: Rent per square metre
sqm_data = df.dropna(subset=['rent','floor_sqm']).copy()
# Remove unrealistic floor areas (less than 10sqm or over 500sqm)
sqm_data = sqm_data[sqm_data['floor_sqm'].between(10, 500)]
# Calculate rent per square metre — a fairer comparison across property sizes
sqm_data['rent_per_sqm'] = (sqm_data['rent'] / sqm_data['floor_sqm']).round(2)
# What is the typical rent per sqm across London?
print(f'Median rent per sqm: £{sqm_data["rent_per_sqm"].median():.2f}/month')
print(f'Mean rent per sqm: £{sqm_data["rent_per_sqm"].mean():.2f}/month')
# Scatter plot: floor area vs monthly rent
# Colour by number of bedrooms
plot_data = sqm_data[sqm_data['bedrooms'].between(1,5) & (sqm_data['rent'] < 8000)]
fig, ax = plt.subplots(figsize=(11, 6))
for beds in [1, 2, 3, 4, 5]:
    sub = plot_data[plot_data['bedrooms'] == beds]
    ax.scatter(sub['floor_sqm'], sub['rent'],
    alpha=0.25, s=8, label=f'{int(beds)} bed')
    ax.set_title('Floor area vs monthly rent — coloured by bedroom count',
    fontsize=13, pad=12)
    ax.set_xlabel('Floor area (sqm)', fontsize=10)
    ax.set_ylabel('Monthly rent (£)', fontsize=10)
    ax.legend(title='Bedrooms', fontsize=8)
    ax.grid(True, alpha=0.2)
    plt.tight_layout()
    plt.savefig('outputs/11_floorarea_vs_rent.png', dpi=150, bbox_inches='tight')
    plt.show()
```

```python
# ANALYSIS 4: Sale price by tenure type
tenure_data = df.dropna(subset=['sale_price','tenure']).copy()
# What tenure types exist?
print('Tenure types and counts:')
print(tenure_data['tenure'].value_counts())
# Median sale price by tenure
price_by_tenure = (tenure_data
.groupby('tenure')['sale_price']
.agg(median='median', mean='mean', count='count')
.sort_values('median', ascending=False)
.reset_index()
)
print('\nMedian sale price by tenure:')
for _, row in price_by_tenure.iterrows():
    print(f" {row['tenure']:15s}: £{row['median']/1000:.0f}k ({row['count']:,} properties)")
    fig, ax = plt.subplots(figsize=(9, 5))
    colours = ['#0F6E56','#4B3FA8','#BA7517','#993C1D']
    bars = ax.bar(price_by_tenure['tenure'],
    price_by_tenure['median'] / 1000,
    color=colours[:len(price_by_tenure)],
    edgecolor='white', width=0.5)
for bar, row in zip(bars, price_by_tenure.itertuples()):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+5,
    f'£{row.median/1000:.0f}k', ha='center', fontsize=9, fontweight='bold')
    ax.set_title('Median estimated sale price by tenure type', fontsize=13, pad=12)
    ax.set_xlabel('Tenure', fontsize=10)
    ax.set_ylabel('Median price (£ thousands)', fontsize=10)
    ax.grid(True, axis='y', alpha=0.3)
    ax.tick_params(axis='x', rotation=15)
    plt.tight_layout()
    plt.savefig('outputs/12_price_by_tenure.png', dpi=150, bbox_inches='tight')
    plt.show()
```
